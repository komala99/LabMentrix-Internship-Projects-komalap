# Exploratory Data Analysis (EDA)

## Restaurant Cost Analysis

Restaurant cost is one of the most important factors that influence customer preferences. In this section, we analyze the cost of restaurants to understand its distribution and identify the minimum, maximum, and average cost.

This analysis helps us understand the pricing patterns of restaurants listed on Zomato.

## Restaurant Cost Analysis

In this section, we analyze the cost of restaurants available in the dataset. The objective is to understand the pricing of restaurants by finding the minimum, maximum, and average costs. We also visualize the cost distribution using a histogram.

Python cannot calculate the average because these are text (strings), not numbers.

This line removes:

₹ symbol
Comma (,)

```python
metadata["Cost"] = metadata["Cost"].replace('[₹,]', '', regex=True)
```

this converts the Cost column into numeric values.

If any value cannot be converted (for example, "Not Available"), it becomes NaN (missing value) instead of causing an error.

```python
metadata["Cost"] = pd.to_numeric(metadata["Cost"], errors="coerce")
```

It calculates the average restaurant cost.

```python
print("Average Cost:", metadata["Cost"].mean())
```

It finds the highest restaurant cost in the dataset.

```python
print("Maximum Cost:", metadata["Cost"].max())
```

It finds the lowest restaurant cost.

```python
print("Minimum Cost:", metadata["Cost"].min())
```

## Distribution of Restaurant Cost Using Histogram

A histogram is a graphical representation of the distribution of numerical data. It groups restaurant costs into different price ranges (bins) and shows the number of restaurants falling within each range.

This visualization helps us understand whether most restaurants are low-cost, medium-cost, or high-cost. It also helps identify the spread of restaurant prices and detect any unusually expensive restaurants (outliers).

```python
plt.figure(figsize=(8,5))
plt.hist(metadata["Cost"].dropna(), bins=20)
plt.title("Distribution of Restaurant Cost")
plt.xlabel("Restaurant Cost (₹)")
plt.ylabel("Number of Restaurants")
plt.show()
```

## Observation

The histogram indicates that the majority of restaurants have costs between ₹300 and ₹800. As the restaurant cost increases, the number of restaurants decreases. Only a few restaurants have costs above ₹2000, indicating that premium restaurants are less common. Therefore, the dataset is dominated by affordable and moderately priced restaurants.

## Top 10 Most Popular Cuisines

In this section, we analyze the cuisines offered by restaurants in the dataset. A cuisine represents the type of food served by a restaurant, such as North Indian, Chinese, South Indian, Italian, etc.

The objective of this analysis is to identify the most common cuisines available on Zomato. This helps us understand customer preferences and the popularity of different food categories.

```python
top_cuisines = metadata["Cuisines"].value_counts().head(10)

print(top_cuisines)
```

```python
plt.figure(figsize=(10,5))
top_cuisines.plot(kind="bar")
plt.title("Top 10 Most Popular Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.show()
```

## Observation

The bar chart shows the top 10 most popular cuisines in the dataset. The combination of **North Indian and Chinese** is the most common cuisine, followed by **North Indian** cuisine. Other cuisine combinations, such as **North Indian, Chinese, Continental** and **Biryani, North Indian, Chinese**, appear less frequently. This indicates that North Indian cuisine, either alone or combined with other cuisines, is the most popular among the restaurants in this dataset.

## Restaurant Timings Analysis

Restaurant timings indicate the operating hours of restaurants. In this section, we identify the most common opening schedules among the restaurants. This helps us understand the availability and operating patterns of restaurants in the dataset.

```python
top_timings = metadata["Timings"].value_counts().head(10)

print(top_timings)
```

```python
plt.figure(figsize=(10,5))
top_timings.plot(kind="bar")
plt.title("Top 10 Restaurant Timings")
plt.xlabel("Timings")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.show()
```

## Observation

The bar chart shows the top 10 most common restaurant operating timings in the dataset. The timing **11 AM to 11 PM** is the most frequently observed operating schedule, indicating that many restaurants remain open throughout the day until late evening. Other common timings include **12 Noon to 12 Midnight** and **12 Noon to 11 PM**. A few restaurants operate until 1 AM or have different schedules on weekends, showing that some restaurants extend their business hours to serve more customers.

## Restaurant Ratings Analysis

Customer ratings are an important indicator of restaurant quality and customer satisfaction. In this section, we analyze the distribution of restaurant ratings given by customers.

This analysis helps us understand whether most restaurants receive high, average, or low ratings.

```python
reviews["Rating"] = pd.to_numeric(reviews["Rating"], errors="coerce")

print(reviews["Rating"].describe())
```

```python
plt.figure(figsize=(8,5))
plt.hist(reviews["Rating"].dropna(), bins=10)
plt.title("Distribution of Restaurant Ratings")
plt.xlabel("Ratings")
plt.ylabel("Number of Reviews")
plt.show()
```

## Observation

The histogram shows the distribution of customer ratings for restaurants. The majority of reviews have ratings of **5**, followed by **4**, indicating that most customers were highly satisfied with their dining experience. Comparatively, fewer reviews have ratings of **1**, **2**, or **3**, suggesting that negative feedback is less common. Overall, the dataset reflects a positive customer perception of the restaurants.

## Review Length Analysis

Customer reviews vary in length depending on the feedback provided. In this section, we calculate the length of each review and analyze its distribution.

This helps us understand whether customers usually write short reviews or detailed reviews.

```python
reviews["Review_Length"] = reviews["Review"].astype(str).apply(len)

reviews["Review_Length"].describe()
```

```python
plt.figure(figsize=(8,5))
plt.hist(reviews["Review_Length"], bins=20)
plt.title("Distribution of Review Length")
plt.xlabel("Number of Characters")
plt.ylabel("Number of Reviews")
plt.show()
```

## Observation

The histogram shows that the majority of customer reviews are short in length, containing a relatively small number of characters. As the review length increases, the number of reviews decreases significantly. Only a few reviews are very long, indicating that most customers prefer to provide brief feedback, while a small number write detailed reviews about their dining experience.
